<h3>Introduction</h3>

<p>Since 1995, the level of crime recorded by the Crime Survey of England (CSEW) has dropped significantly but surveys to evaluate people’s perceptions of crime have shown that individuals still perceive crime to be rising (Flatley, 2017a). In the year 2017, 60% of adults believed that crime rate was rising but this was an improvement from the year 2009 when about 85% of adults believed the same (Flatley, 2017a). In general, the total number of individuals who believe that they are vulnerable to attacks such as mugging, burglary, and physical attacks is slowly reducing but the general perception of a lack of safety prevails.</p>

<p>According to Gray et al (2008), the fear of crime often affects the behavior of individuals and their perceptions towards issues such as race and class within society. For instance, if Black men are arrested occasionally for burglary or physical assaults, people may be inclined to think that crime is carried out only by people of color (Gray et al., 2008). The fear and worry about crime can affect people’s choice of where they live. In general, individuals avoid neighborhoods that have been marked as unsafe and, it is common to find the housing rates in these areas being lower than the average.</p> 

<p>Since fear and worry about crime affect various aspects of people’s lives, the following report is aimed at evaluating the nature and motivations behind crime within England and Wales. The correlation between the factors influencing the crime rates with people’s perceptions is also investigated for a better understanding of the subject.</p>

<p><b>Relevant Variables</b></p>

<p>In the study by CSEW, the participants were from different age groups and races. Both male and female respondents were included and the participants were from both rural and urban settings. By including the number of years lived by the individuals in these areas, the credibility and relevance of the observed responses would be evaluated. </p> 

<p>Some of the notable variables in the study are the level of worry about being mugged or robbed, the level of worry about having one’s home being broken into, the level of worry about being physically attacked by strangers, and the level of worry about being alone in home at night. In this study, the variables analyzed are the level of fear in relation to crime types, the fear of being raped, worry about having things stolen from one’s car, the fear of being physically attacked by strangers, and the worry about being attacked due to race.</p> 

<p><b>Fear of Having Things Stolen from One’s Car</b></p>

<p>From the analysis conducted, it is clear that a majority of the population in England and Wales have no fear of having things stolen from their vehicles. 50% are not very worried while only 5% are very worried about it. According to Flately (2017b), vehicle-related thefts range from the lifting of things from people’s cars through door windows or the stealing of vehicle accessories and vehicles in general. The author notes that some of the motivations behind this behavior are a lack of awareness among drivers and vehicle owners, poverty which leads to homelessness and street children who are ready to commit such crimes for a living, poor vehicle security, and inadequate support from the police. Since the year 1981, the trend in vehicle-related thefts has been declining, thus the trend seen in the pie chart presented in this report (Flately, 2017b).</p>

<p><b>The Fear of Being Raped</b></p>

<p>According to Khadr et al. (2018), rape cases within England and Wales have been declining since 1993 due to several reasons. First, the government has set up strict punishment strategies for those who commit such crimes, there has been a rise in awareness about rape among the populations of the two nations, and various non-governmental organizations are now available to offer help whenever individuals feel threatened. The authors note that the people most likely to fall victims to rape are the physically disabled, mentally unstable, or those without basic training to prevent rape.</p> 

<p>From the pie chart presented in this analysis, it can be seen that about 49% of the respondents were not worried about rape while only about 15% were worried. This can be attributed to the fact that most men are not usually vulnerable to rape, leaving only women to worry about it. Among these, a great percentage of them live in safe neighborhoods or have some basic form of training or education to protect them from being victims of rape. Thus, it can be concluded that the percentage of individuals worried about rape are those without proper education on rape prevention or those living in unsafe neighborhoods.</p>

<p><b>Fear of Being Physically Attacked</b></p>

<p>From the analysis presented, about 47% of the respondents were not very worried about being attacked physically while about 8% were very worried. Generally, about 26% of the respondents had some level of worry about physical attacks and this can be attributed to several reasons. First, some may be physically disabled. This group normally perceives physical attacks as a real worry because of their inability to properly defend themselves (Lorenc et al., 2014). Second, individuals who have been attacked in the past by a particular group of people within society often harbor the fear of being attacked in the future by similar people (Lorenc et al., 2014). Finally, women generally fear that they may be attacked by rough-looking men in the streets for no rational reason (Lorenc et al., 2014).</p>

<p><b>Worry About Being Attacked Due to Race</b></p>

<p>In the modern world, international communication and transport have enabled the interaction between people from different worlds. That has facilitated cultural exchanges and the sharing of beliefs, increasing cohesion within societies. While racial injustices were dominant in the past century, they have significantly declined in the 21st century and continue to decline as people interact increasingly. Thus, it can be seen from the pie chart presented that only about 10% of the respondents were worried about being attacked based on race. A majority, 59%, were not worried at all and, that can be attributed to globalization. The slight worry observed in about 28% of the respondents can be attributed to past experiences with racial discrimination or injustices this group had observed being carried out on other people.</p>

<p><b>Percentage of People Who Have Been Victims of Crime Before</b></p>

<p>From the bcsvictim pie chart, it can be seen that about 84% of the respondents had never been victims of fear before the study. Only about 16% of them had experienced some form of crime and Lorenc et al. (2014) highlight that this can skew people’s perceptions of crime. Individuals who have experienced criminal injustices are more fearful of reoccurrence than those who have never experienced any form of crime (Lorenc et al., 2014). Therefore, it can be concluded that although the crime rates within England and Wales have declined since the beginning of this decade, there are still people who harbor irrational fears of crime due to past experiences. As respondents, such individuals would be part of the group that is afraid of being attacked physically or on a racial basis.</p>


<p><b>Contextual Variables</b></p>

<p>In general, it can be observed from the analyzed data that about 50% of the populations in each category of a variable are not afraid of crime within their neighborhoods. This can be taken to be representative of the whole population. The observed lack of fear, in general, can be attributed to the rise in efforts by governments to improve security within their nations. In England and Wales, there have been efforts to establish various departments within the police force for dealing with different crimes (Griffiths & Norris, 2020). This relieves the police force of the number of criminal cases to be investigated and solved, making justice delivery timely and more efficient. The collection of data on past criminal activities within England and Wales has enabled the government to map out the regions most vulnerable to criminal activities while relating them with possible causes (Griffiths & Norris, 2020). Through addressing these causes, the overall crime rates have declined.</p>

<p>The median percentage of worry about crime is 40%. This is representative of the percentage of the population within England and Wales which has some degree of fear about crime. As mentioned earlier, the fear is correlated with the neighborhood within which individuals live, their marital status, and race. Wealthy individuals can afford to live in safe neighborhoods where crime is lower but for the majority of the middle and lower class individuals, the choice of the neighborhood to live in is limited. Also, people who are married have less fear of having their homes broken into compared to those who are not. Having a partner creates a sense of security and, therefore, improving one’s perception of possible criminal attacks (Griffiths & Norris, 2020). Finally, respondents of color are more likely to express the fear of being attacked on a racial basis than their counterparts. Although the fear may seem rational, Flately (2017a) notes that the level of race-based discrimination cases has declined significantly in the 21st century.</p>
