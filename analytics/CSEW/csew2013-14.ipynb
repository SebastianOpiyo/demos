### 1. What is the level of fear in relation to crime types?

```python
# Plotting a pie chart ranking levels of feat/worry

# For wfromcar
crimedata['wfromcar'].value_counts().plot(kind = 'pie', colors=['green', 'gold', 'black', 'pink','orange','blue'], figsize = (8, 8), autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.title('Pie Chart for "wfromcar" levels', fontsize = 20)
plt.xlabel('Fear levels')
plt.ylabel('')
plt.show()

```

```python
# For wraped
crimedata['wraped'].value_counts().plot(kind = 'pie', colors=['green', 'gold', 'black', 'pink','orange','blue'], figsize = (8, 8), autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.title('Pie Chart for "wraped" levels', fontsize = 20)
plt.xlabel('Fear levels')
plt.ylabel('')
plt.show()



```

```python
# For 'wattack'
crimedata['wattack'].value_counts().plot(kind = 'pie', colors=['green', 'gold', 'black', 'pink','orange','blue'], figsize = (8, 8), autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.title('Pie Chart for "wattack" levels', fontsize = 20)
plt.xlabel('Fear levels')
plt.ylabel('')
plt.show()

```

```python
# For 'wraceatt'
crimedata['wraceatt'].value_counts().plot(kind = 'pie', colors=['green', 'gold', 'black', 'pink','orange','blue'], figsize = (8, 8), autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.title('Pie Chart for "wraceatt" levels', fontsize = 20)
plt.xlabel('Fear levels')
plt.ylabel('')
plt.show()

```

### 2. What fear or worry scales the highest or lowest ?

```python
# For 'worryx'
crimedata['worryx'].value_counts().plot(kind = 'bar', color=['green', 'gold', 'black', 'pink','orange','blue'], figsize = (8, 8))
plt.title('Pie Chart for "worryx" levels', fontsize = 20)
plt.xlabel('Fear levels')
plt.ylabel('')
plt.show()

```

```python
# For bcsvictim
crimedata['bcsvictim'].value_counts().plot(kind = 'pie', colors=['green', 'gold', 'black', 'pink','orange','blue'], figsize = (8, 8), autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.title('Pie Chart for "bcsvictim" levels', fontsize = 20)
plt.xlabel('Victims of crime.')
plt.ylabel('')
plt.show()
```
